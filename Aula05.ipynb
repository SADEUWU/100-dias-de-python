## Classes

```python
class Circle(object):

    pi = 3.142

    def  __init__(self, radius):
        self.radius = radius

    def area(self):
        return self.pi * (self.radius ** 2)
```

```python
a = Circle(32)
print(a.area())
print(a.pi)
a.pi = 50
print(a.area())
b = Circle(32)
print(b.area())
```

```python
class SuperClass(object):
    superpowers = []

    def __init__(self, name):
        self.name = name

    def add_superpower(self, power):
        self.superpowers.append(power)
```

```python
peter = SuperClass("Homem Aranha")

print(peter.name)

peter.add_superpower("Se agarrar a prédios")
print(peter.superpowers)
```

```python
clark = SuperClass("Superman")
print(clark.name)
clark.add_superpower("Voar")
print(clark.superpowers)
```

## Medidas

```python
values = [1, 2, 7, 12, 13]
mean = sum(values)/len(values)
di = sum([value-mean for value in values])
print("Média:", mean) 
print("Soma dos desvios em relação a média:", di)
```

## Distribuição Gaussiana

```python
xAxis = np.arange(-3, 3, 0.001)
yAxis = norm.pdf(xAxis, 0, 1)
plt.plot(xAxis, yAxis)
plt.show()
```

```python
data = 5 * randn(1000) + 50
plt.hist(data, bins=100)
plt.show()
```

```python
print('Média: %.3f' % np.mean(data))
print('Mediana: %.3f' % np.median(data))
print('Moda:', stats.mode(data))
```

## Distribuição de Poisson

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
dp = np.random.poisson(lam=3, size=(1000))
plt.hist(dp)
plt.show()
```

```python
sns.distplot(dp)
plt.show()
```

```python
ax = sns.distplot(np.random.poisson(lam=3, size=20000), bins=27, kde=False)
ax.set(xlabel="Mercadorias", ylabel="Unidades Vendidas")
plt.show()
```

## Distribuição Binomial

```python
db = np.random.binomial(n=10, p=0.5, size=1000)
plt.hist(db)
plt.show()
```

```python
sns.distplot(db)
plt.show()
```
