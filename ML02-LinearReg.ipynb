# Modelo "Regresso Linear"

Algortmo Supervisionado destinado a predies

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn import linear_model

%matplotlib inline
```

## Populao Brasileira

```python
df = pd.read_csv('bases/PopBrasil.csv')
df.head(5)
```

```python
df.tail(5)
```

## Plotar o grfico

```python
plt.xlabel('Ano')
plt.ylabel('Quantidade da Populao')
plt.scatter(df.Ano, df.Populacao, color='red', marker='+')
```

```python
reg = LinearRegression()
reg.fit(df[['Ano']], df.Populacao)
prev = reg.predict([[2024]])
print("Previso 2024 : %d" % prev)
```

```python
plt.xlabel('Ano')
plt.ylabel('Quantidade da Populao')
plt.scatter(df.Ano, df.Populacao, color='red', marker='+')
plt.plot(df.Ano, reg.predict(df[['Ano']]), color='blue')
```

## Como  realizado o Clculo?
Visitar a pgina: https://www.mathsisfun.com/algebra/linear-equations.html

Frmula: y = mx + b

```python
m = reg.coef_[0]
```

```python
b = reg.intercept_
```

```python
prev2024 = m * 2024 + b
print("Previso 2024 : %d" % prev2024)
```
